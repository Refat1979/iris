```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt 
```

```python
sns.set(style = "white", color_codes = True)
```

```python
iris = pd.read_csv("../Iris.csv")
```

```python
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
iris.plot(kind ='scatter', x='SepalLengthCm', y= 'SepalWidthCm')
```

```python
sns.jointplot(x='SepalLengthCm', y= 'SepalWidthCm', data = iris, size =5)
```

```python
sns.FacetGrid(iris, hue = "Species") .map(plt.scatter, 'SepalLengthCm', 'SepalWidthCm') .add_legend()
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris, hue = "Species")
```

```python
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
```

```python
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris,hue = "Species")
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray", hue = "Species")
```

```python
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, hue = "Species")
```

```python
sns.FacetGrid(iris, hue="Species") .map(sns.kdeplot, "PetalLengthCm") .add_legend()
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")
```

```python
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6) )
```
